```python
%matplotlib notebook
import pyscreenshot as ImageGrab
from imageio import imread, imwrite
from pylab import *
import os
```

```python
screenshot_dir = os.path.abspath('../test_data/ridiculous6/screenshot/')
```

```python
filepath = os.path.join(screenshot_dir, '{}.png'.format(int(np.random.random()*100000)))
image = np.array(ImageGrab.grab())[:,1920:2*1920,:]

imwrite(filepath, image)

figure()
imshow(image)
show()
```
