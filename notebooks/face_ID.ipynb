## 2. Import Identity Images

```python
im_id_path = '../train_data/FaceID/Parks&Rec/nick_offerman/nick_offerman'
identity_images = list()
fig_input = plt.figure()
figlab = 'abcd'
for i in range(1,5):
    identity_images.append(imageio.imread(im_id_path + '{}.JPG'.format(i)))
    fig_input.add_subplot(2,2,i)
    plt.imshow(identity_images[-1])
    plt.xlabel('({})'.format(figlab[i-1]))
      
plt.tight_layout()
plt.suptitle('Nick Offerman')
plt.subplots_adjust(top=0.90)
plt.show()
```
