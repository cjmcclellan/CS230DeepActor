```python
%matplotlib notebook
import os
import pickle
from pylab import *
```

```python
model_dir = os.path.abspath('../models/softmax/ridiculous6/1915/')
```

```python
model_perf = pickle.load(open(os.path.join(model_dir, 'model_performance.pkl'), 'rb'))
```

```python
# Get description of object
print(model_perf.keys())
```

```python
train_loss = model_perf['train_loss']
validation_loss = model_perf['validation_loss']
train_accuracy = model_perf['train_accuracy']
validation_accuracy = model_perf['validatio_accuracy']
```

```python
figure()
semilogy(train_loss)
semilogy(validation_loss)
xlabel('Epochs')
ylabel('Cross Entropy Loss')
show()
print('Minimum Training Loss:', min(train_loss))
print('Minimum Validation Loss:', min(validation_loss))
```

Realistically, the validation error stops changing around epoch 2500. This will be the model I restore for testing. More training will overfit. This is our regularization technique

```python
figure()
semilogx(train_accuracy)
semilogx(validation_accuracy)
xlabel('Epochs')
ylabel('Accuracy')
ylim([0.6, 1.01])
show()
print('Max Training Accuracy', max(train_accuracy))
print('Max Validation Accuracy', max(validation_accuracy))
```
